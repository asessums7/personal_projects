```python
module_ids = []

for modules in learning_space_soup.find_all('li', id=True):  
    modules = modules.get('id')
    module_ids.append(modules)

module_ids
```

```python
learning_space_links = []

for songlink in module_ids:
    learning_space_url = "https://learning.peacecorps.gov/mod/book/view.php?id=" + songlink
    learning_space_links.append(learning_space_url)
    print(learning_space_links)
```

```python
def distinct(links):
    '''Function to de-dup list'''
    list(dict.fromkeys(links))
    print(links)
```

```python
distinct_learning_space_links = distinct(learning_space_links)
distinct_learning_space_links
```
